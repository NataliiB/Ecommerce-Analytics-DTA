```python
df = df_raw.copy()
```

```python
df['Purchase Date'] = pd.to_datetime(df_raw['Purchase Date'], errors='coerce')
df['Purchase Date'].dtype
```

```python
df['Discount Name'] = df['Discount Name'].fillna('No Discount')
```

```python
df['is_discount_overflow'] = df['Discount Amount (INR)'] >= df['Gross Amount']
df['is_discount_overflow'].value_counts()
```

```python
numeric_columns = df.select_dtypes(include='number')
plt.figure(figsize=(10,8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation matrix of financial indicators')
plt.show()
```

```python
gender_preferences = df.groupby(['Gender', 'Product Category']).size().unstack()
sns.heatmap(gender_preferences, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Category popularity by gender')
plt.show()
```

```python
sns.scatterplot(
    x='Gross Amount',
    y='Net Amount',
    hue='is_discount_overflow',
    data=df
)
```

```python
sns.histplot(df[df['is_discount_overflow']]['Gross Amount'])
```

```python
df_negative = df[df['Net Amount'] < 0].copy()
negative_by_month = df_negative.resample('ME', on='Purchase Date').size()

fig, ax = plt.subplots(figsize=(12, 6))
negative_by_month.plot(kind='bar', ax=ax, color='salmon')

new_labels = [d.strftime('%b-%Y') for d in negative_by_month.index]

ax.set_xticklabels(new_labels, rotation=90, ha='right')

plt.title('Distribution of anomalous values â€‹â€‹(Net Amount < 0) by month', fontsize=14)
plt.ylabel('Number of transactions')
plt.xlabel('Month and Year')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

plt.show()
```

Final Exploratory Data Analysis (EDA) Summary
Project Overview: The dataset consists of 55,000 e-commerce transactions with 13 features, covering the period from 2019 to 2024.

1. Data Structure and Integrity
Data Types: The dataset contains 5 numerical and 7 categorical features. The Purchase Date was successfully converted to datetime64 for accurate time-series analysis.

Data Completeness: No duplicate records were found. The Discount Name column contained 27,585 missing values, which were filled with 'No Discount' as they consistently corresponded to a zero discount amount.

2. Customer Demographics & Segmentation
Gender Distribution: The customer base is remarkably balanced among Female (33.55%), Other (33.54%), and Male (32.9%). This indicates a highly inclusive market appeal across all gender identities.

Age Groups: The primary audience is the 25-45 age group (40.02%), followed by 18-25 year olds (29.87%). Together, customers aged 18 to 45 drive nearly 70% of all transactions, identifying them as the core economic engine for the platform.

Geographic Concentration: Sales are concentrated in major urban hubs, with Mumbai (20.36%) and Delhi (19.63%) leading the revenue share. Nearly 55% of all orders originate from the top four cities (Mumbai, Delhi, Bangalore, and Hyderabad).

3. Financial Metrics & High-Value Customers
Key Stats: The average transaction value (Gross Amount) is 3,012.94, with a maximum reaching 8,394.83.

Product Categories: Electronics (34.58%) and Clothing (22.02%) are the dominant categories, representing over half of the total transaction volume.

VIP Segment: A specific segment of 52 transactions with values exceeding 8,000 was identified. While statistically classified as outliers, these are logically valid records representing High-Value Customers, essential for premium marketing analysis.

4. Anomalies & Technical Logic Errors
The "Overflow" Issue: Identified 613 transactions (1.1%) where the Discount Amount exceeds the Gross Amount, resulting in a negative Net Amount.

Systemic Patterns: These errors occur consistently over time. Over 64% of anomalies are linked to Credit Card (36.7%) and Debit Card (27.4%) payments, suggesting a technical conflict between automated bank-side discounts and store promotions.

Data Strategy: To maintain reporting accuracy, an is_overflow_discount flag was added. This allows for the exclusion of "negative" revenue from financial dashboards in Tableau without losing the original transaction data.
